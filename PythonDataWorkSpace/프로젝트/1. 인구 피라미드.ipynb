# 1. 인구 피라미드


```python
import pandas as pd
file_name = '202103_202103_인구현황.xlsx'
df_m= pd.read_excel(file_name, skiprows=3, index_col='행정기관', usecols ='B, E:Y')
df_m
df_w= pd.read_excel(file_name, skiprows=3, index_col ='행정기관', usecols = 'B,AB:AV')
#위에 3개만 가져와
df_m.head(3)
```

```python
#string -> 숫자로 만들기 : 1,195,951 -> 1195951
df_m.iloc[0]= df_m.iloc[0].str.replace(',','').astype(int)
df_m
```

```python
df_m.iloc[0]
```

```python
df_w.iloc[0]= df_w.iloc[0].str.replace(',','').astype(int)
df_w.head(3)
```

# 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 깨짐 해결
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) #//1000 은 단위가 1000명
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2021 대한민국 인구 피라미드')
plt.savefig('2021_인구피라미드.png', dpi=100)
plt.show()
```
