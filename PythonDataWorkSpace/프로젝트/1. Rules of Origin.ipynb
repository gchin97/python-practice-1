# 1. 프로젝트 설명 

## 프로젝트명: 세계 무역량 데이터를 이용한 FTA 이후 개발도상국의 무역량 영향 분석 
### 개요 및 진행 배경: FTA로 인하여 개발도상국의 무역량 감소 및 산업을 파악하고 이에 대한 FTA의 개선점이 필요하다는 제언을 하고자 진행


```python
import pandas as pd
df_ITrade= pd.read_excel('International_Trade_data_TradeVolume.xlsx', skiprows=3, usecols ='A:AC')
df_ITrade = df_ITrade.fillna(0)
df_ITrade[df_ITrade.columns[2:]]=df_ITrade[df_ITrade.columns[2:]].replace(',','',regex=True).astype('int64')
# df_ITrade['Trade_Value']= df_ITrade.apply(lambda x: "{:,}".format(x['Trade_Value']), axis=1)
df_ITrade
df_ITrade.columns = df_ITrade.columns.astype(str)
df_ITrade.set_index('Country')
```

```python
df_NumTrade= pd.read_excel('regional_agreements_byCountry.xlsx')
df_NumTrade
```
