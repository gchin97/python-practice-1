# 04. style

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 깨짐 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle ='None')
```

```python
plt.plot(x,y, marker='v', linestyle ='dashed')
```

```python
plt.plot(x,y, marker='v', markersize= 10, markeredgecolor= 'red')
```

```python
plt.plot(x,y, marker='v', markersize= 10, markeredgecolor= 'red', markerfacecolor='yellow')
```

# 선스타일

```python
plt.plot(x,y,linestyle=':')
```

```python
plt.plot(x,y,linestyle='--')
```

```python
plt.plot(x,y,linestyle="-.")
```

```python
plt.plot(x,y,linestyle='-')
```

# 색깔

```python
plt.plot(x,y,color="red")
```

```python
plt.plot(x,y,color="b")
```

```python
plt.plot(x,y,color="g")
```

## 포맷

```python
plt.plot(x,y,'ro--') #컬러, 마커, 라인스타일
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go') # color = g marker = 0 linestyle='-'
```

# 축약어

```python
plt.plot(x,y,marker ='o', mfc='red', ms=10, mec = 'blue', ls = ':')
```

```python
plt.plot(x,y,marker ='o', mfc='red', ms=10, mec = 'blue', alpha=0.3)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,4), dpi=200) #dots per inch 해상도를 정의할 수 있음
plt.plot(x,y)
```

# 그래프 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```
