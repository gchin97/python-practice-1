# 02. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'AppleGothic' 
matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글폰트 사용시 마이너스 글자 깨짐 해결
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은 선 그래프')
plt.title('꺾은 선 그래프', fontdict ={'family': 'HYGungSo-Bold','size': '20'}) # 개별 폰트 설정이 가능함
```

```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('y축')
```

```python
plt.plot(x,y)
plt.xlabel('X축', color = 'red')
plt.ylabel('y축', color = '#00aa00')
```

```python
#축의 이름 이동시키기 
plt.plot(x,y)
plt.xlabel('X축', color = 'red', loc = 'right') #left, center, right
plt.ylabel('y축', color = '#00aa00', loc = 'top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
