## 데이터 제외하기 dropna

```python
import pandas as pd 
df= pd.read_excel("score.xlsx", index_col="지원번호")
df
```

```python
df.dropna() #NaN 이 포함되어 있는 row 모든 데이터 삭제 
```

```python
df.dropna(inplace=True)
df
```

```python
import pandas as pd 
df= pd.read_excel("score.xlsx", index_col="지원번호")
df
```

```python
axis: index or columns # index는 row 기준으로 삭제, col 기준으로 삭제 
how: any or all  # any 일시 데이터 중 하나라고 NaN이 있다면 지우는 것, all 이라면 NaN이 삭제됨 
```

```python
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='index', how = 'any') #NaN이 하나라도 있는 로우 삭제 
```

```python
df.dropna(axis="columns") # NaN이 하나라도 있는 컬럼 삭제 
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how="all") #데이터 전체가 NaN 인 경우에만 Column 삭제, SW 특기는 두개의 데이터만 nan 이기 때문에 Column이 안 지워짐
```
